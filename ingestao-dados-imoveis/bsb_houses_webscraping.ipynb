```python
import requests
import pandas as pd

from bs4 import BeautifulSoup as bs
```

```python
# url
url = 'https://www.vivareal.com.br/venda/distrito-federal/brasilia/apartamento_residencial/?pagina={}'
```

```python
i = 1
```

```python
#request to URL
r = requests.get(url.format(i))
```

```python
# BeautifulSoup object
soup = bs(r.text, 'html.parser')
```

```python
# houses per page
houses = soup.find_all('a', class_="property-card__content-link js-card-title")

# total qty of houses available in this city
qty_houses = float(soup.find('strong', class_ = "results-summary__count js-total-records").text.replace('.', ''))
```

```python
len(houses)
```

```python
# total pages to scrap
qty_houses/len(houses)
```

```python
house = houses[0]
```

```python
### features ###
# new dataframe
df_houses = pd.DataFrame(columns = ['description',
'address',
'area',
'qty_bedrooms',
'qty_bathrooms',
'parking_spaces',
'price',
'condo',
'hlink'])

# defining loop for pagination
i = 0
```

```python
# web scraping pagination 
while qty_houses > df_houses.shape[0]:
    i += 1
    print(f'page: {i} \t\t qty_houses: {df_houses.shape[0]}')
    
    page = requests.get(url.format(i)) #different pages defined by i iteration
    soup = bs(page.text, 'html.parser') #soup object
    # houses per page
    houses = soup.find_all('a', class_="property-card__content-link js-card-title")

    for house in houses:
        # description
        try:
            description = house.find('span', class_ = 'property-card__title js-cardLink js-card-title').text.strip() # strip to get rid of empty spaces in the begging
        except:
            description = None
        # address
        try:
            address = house.find('span', class_ = "property-card__address").text.strip() # strip to get rid of empty spaces in the begging
        except:
            address = None
        # area
        try:
            area = house.find('span', class_ = "property-card__detail-value js-property-card-value property-card__detail-area js-property-card-detail-area").text.strip() # strip to get rid of empty spaces in the begging
        except:
            area = None
        # qty of bedrooms
        try:
            qty_bedrooms = house.find('li', class_ = 'property-card__detail-item property-card__detail-room js-property-detail-rooms').span.text.strip() # strip to get rid of empty spaces in the begging
        except:
            qty_bedrooms = None
        # bathrooms
        try:
            qty_bathrooms = house.find('li', class_ = 'property-card__detail-item property-card__detail-bathroom js-property-detail-bathroom').span.text.strip() # strip to get rid of empty spaces in the begging
        except:
            qty_bathrooms = None
        # parking spaces
        try:
            parking_spaces = house.find('li', class_ = 'property-card__detail-item property-card__detail-garage js-property-detail-garages').span.text.strip() # strip to get rid of empty spaces in the begging
        except:
            parking_spaces = None
        # price value
        try:
            price = house.find('div', class_ = 'property-card__price js-property-card-prices js-property-card__price-small').p.text.strip() # strip to get rid of empty spaces in the begging
        except:
            price = None
        # condo
        try:
            condo = house.find('strong', class_ = 'js-condo-price').text.strip() # strip to get rid of empty spaces in the begging
        except:
            condo = None
        # house link
        try:
            hlink = 'https://www.vivareal.com.br/' + house['href']
        except:
            hlink = None
        
        # adding scrapped data into last line of the dataframe
        df_houses.loc[df_houses.shape[0]] = [description,
                                             address,
                                             area,
                                             qty_bedrooms,
                                             qty_bathrooms,
                                             parking_spaces,
                                             price,
                                             condo,
                                             hlink]
```
